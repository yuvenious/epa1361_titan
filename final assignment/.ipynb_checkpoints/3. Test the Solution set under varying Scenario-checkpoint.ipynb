```python
import time
import numpy as np
import networkx as nx
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from ema_workbench import (Model, ScalarOutcome, MultiprocessingEvaluator, Policy, Scenario)

from ema_workbench.em_framework.evaluators import perform_experiments
from ema_workbench.em_framework.samplers import sample_uncertainties
from ema_workbench.em_framework.optimization import EpsilonProgress, HyperVolume


from ema_workbench.util import ema_logging, save_results, load_results

from ema_workbench.analysis import prim, pairs_plotting, parcoords

from problem_formulation import get_model_for_problem_formulation
ema_logging.log_to_stderr(ema_logging.INFO)
```

```python
dike_model = get_model_for_problem_formulation(1)
```

```python
# Load Policy
results = pd.read_csv("SolsRefScen.csv", index_col=0)
policies = []
for i in range(len(results)):
    pol_dict = results.iloc[:,:11].loc[i].to_dict()
    policy = Policy('Policy %s'%i, **pol_dict)
    policies.append(policy)
```

```python
n_scenarios = 100

start = time.time()
with MultiprocessingEvaluator(dike_model) as evaluator:
    results = evaluator.perform_experiments(n_scenarios, policies = policies)
    end = time.time()
print(end - start)
save_results(results, "3obj4pol_results.tar.gz")
```

```python
results = load_results("3obj4pol_results.tar.gz")
```

```python
fig, ax = pairs_plotting.pairs_scatter(results, group_by="policy")
fig.set_size_inches(10,10)
```

```python
inputs = pd.DataFrame(results[0])
outcomes = pd.DataFrame(results[1])
```

### Define Failure

```python
q = 0.8
fail = (outcomes > outcomes.quantile(q)).apply(lambda x: "fail" if any(x) else "non-fail", axis = 1)
inputs["fail"] = fail.values
```

```python
sns.countplot("fail", data=inputs)
```

```python
params = list(dike_model.uncertainties.keys())
params.append("policy")
params.append("fail")
```

```python
import itertools

ncols=2
nrows=6
fig, axes = plt.subplots(ncols = ncols, nrows = nrows,
#                          sharex = True,
                         figsize = (5*ncols,5*nrows))
locs = list(itertools.product(range(nrows),range(ncols)))

params = list(dike_model.uncertainties.keys())
params = [param for param in params if ("Brate" not in param) & ("discount" not in param)]

for i,param in enumerate(params):
    loc=locs[i]
    ax=axes[loc]
    sns.violinplot("fail", param, data = inputs, ax=ax, legend=True,
                   hue="policy"
                  )
    ax.set_xlabel("")
    ax.set_xticklabels(inputs["fail"].unique(),fontsize=15)
    ax.set_ylabel("")
    ax.set_title(param)
fig.tight_layout()
fig.suptitle("How Policy can fail/success under varying Uncertainties", y=1.02, fontsize = 15)
```

#### Vulnerability for Categorical Uncertainties

```python
params = list(dike_model.uncertainties.keys())
params = [param for param in params if ("Brate" in param) | ("discount" in param)]
for param in params:
    g = sns.factorplot(x=param, hue="fail", col="policy", data=inputs, kind="count", palette="coolwarm")
    g.set_xlabels("")
    g.fig.suptitle(param, y=1.05, fontsize=15)
```

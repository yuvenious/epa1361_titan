```python
outcomes = pd.DataFrame(outcomes)
outcomes = outcomes.join(pd.DataFrame(experiments)[["scenario_id", "policy"]])
outcomes.head()
```

```python
regrets = outcomes.pivot(index="scenario_id", columns="policy")
regrets.head()
```

```python
overall_regret=pd.DataFrame()
regret = {}
for outcome in regrets.columns.levels[0]:
    max_regret = regrets.loc[:, (outcome, slice(None))]
    ref = max_regret.apply(lambda x: x.min(), axis=1)[:,None]
    max_regret = (ref - max_regret).abs()
    regret[outcome] = max_regret.max().values
    overall_regret = pd.concat([overall_regret, max_regret], axis=1)
regret = pd.DataFrame(regret, index=regrets.columns.levels[1])
regret
```

```python
sns.heatmap(regret/regret.max(), annot=True, cmap="coolwarm")
```

```python
data = regret/regret.max()
limits = parcoords.get_limits(data)
paraxes = parcoords.ParallelAxes(limits)
paraxes.plot(data)
```

```python
import numpy as np
import functools

def robustness(threshold, data):
    return np.sum(data<=threshold)/data.shape[0]

thresholds = [1.06273272e+09, 7.62009753e-01, 1.66523363e+08]

# def func1(data):
#     threshold = thresholds[0]
#     return np.sum(data<=threshold)/data.shape[0]


# def func2(data):
#     threshold = thresholds[1]
#     return np.sum(data<=threshold)/data.shape[0]


# def func3(data):
#     threshold = thresholds[2]
#     return np.sum(data<=threshold)/data.shape[0]


func1 = functools.partial(robustness, thresholds[0])
func2 = functools.partial(robustness, thresholds[1])
func3 = functools.partial(robustness, thresholds[2])
```

```python
# from robust_func import *

robust_funcs = [func1, func2, func3]
robust_funcs
```

```python
ncols=3
nrows=1

fig, axes = plt.subplots(ncols=ncols, nrows=nrows,
                         figsize=(ncols*4,nrows*4),
                        sharex=True, sharey=True)

if any([ncols == 1, nrows == 1]):
    locs = range(max(ncols,nrows))
else:
    locs=list(itertools.product(range(nrows), range(ncols)))

robustness_funcs = dict(zip(outcomes.columns, robust_funcs))
j = 0
for outcome, value in outcomes.items():
    loc = locs[j]
    ax = axes[loc]
    for policy in np.unique(experiments['policy']):
        logical = experiments['policy'] == policy
        data = value[logical]
        
        robustness = []
        for i in range(1, data.shape[0]):
            robustness.append(robustness_funcs[outcome](data[:i]))
        ax.plot(robustness, label=policy)
    ax.set_xlabel("# experiments")
    ax.set_ylabel(outcome)
    ax.legend()
    ax.grid()
    j += 1
fig.tight_layout()
```

```python
frac_names = np.repeat("fraction ", outcomes.shape[1]).astype(object) + outcomes.columns
var_names = outcomes.columns
```

```python
robustness_functions = []
for i, var_name in enumerate(var_names):
    frac_name = frac_names[i]
    function = robust_funcs[i]
    robust_func = ScalarOutcome(name=frac_name, kind=ScalarOutcome.MINIMIZE,
              variable_name=var_name, function=function)
    robustness_functions.append(robust_func)
```

### Define Failure

```python
q = 0.8
fail = (outcomes > outcomes.quantile(q)).apply(lambda x: "fail" if any(x) else "non-fail", axis = 1)
inputs["fail"] = fail.values
```

```python
import itertools

ncols=2
nrows=6
fig, axes = plt.subplots(ncols = ncols, nrows = nrows,
#                          sharex = True,
                         figsize = (5*ncols,5*nrows))
locs = list(itertools.product(range(nrows),range(ncols)))

params = list(dike_model.uncertainties.keys())
params = [param for param in params if ("Brate" not in param) & ("discount" not in param)]

for i,param in enumerate(params):
    loc=locs[i]
    ax=axes[loc]
    sns.violinplot("fail", param, data = inputs, ax=ax, legend=True,
                   hue="policy"
                  )
    ax.set_xlabel("")
    ax.set_xticklabels(inputs["fail"].unique(),fontsize=15)
    ax.set_ylabel("")
    ax.set_title(param)
fig.tight_layout()
fig.suptitle("How Policy can fail/success under varying Uncertainties", y=1.02, fontsize = 15)
```

#### Vulnerability for Categorical Uncertainties

```python
params = list(dike_model.uncertainties.keys())
params = [param for param in params if ("Brate" in param) | ("discount" in param)]
for param in params:
    g = sns.factorplot(x=param, hue="fail", col="policy", data=inputs, kind="count", palette="coolwarm")
    g.set_xlabels("")
    g.fig.suptitle(param, y=1.05, fontsize=15)
```

```python
import numpy as np
import networkx as nx
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from ema_workbench import (Model, MultiprocessingEvaluator, ScalarOutcome, Policy, Scenario)

from ema_workbench.em_framework.evaluators import perform_experiments
from ema_workbench.em_framework.samplers import sample_uncertainties
from ema_workbench.em_framework.optimization import EpsilonProgress, HyperVolume
from ema_workbench.analysis import parcoords

from ema_workbench.util import ema_logging
import time

from problem_formulation import get_model_for_problem_formulation
ema_logging.log_to_stderr(ema_logging.INFO)
```

### 2. Search for a set of Optimal Solutions under Ref_Scenario

```python
dike_model = get_model_for_problem_formulation(3)

# 1. Determine a set of certain values
reference_values = {'Bmax': 175,
                    'Brate': 1.5,
                    'pfail': 0.5, 
                    'discount rate': 3.5,
                    'ID flood wave shape': 4}

# 2. Set the values as a Reference Scenario
scen1 = {}
for key in dike_model.uncertainties:
    name_split = key.name.split('_')
    if len(name_split) == 1:
        scen1.update({key.name: reference_values[key.name]})
    else:
        scen1.update({key.name: reference_values[name_split[1]]})

ref_scenario = Scenario('reference', **scen1)
```

##### If Dike Investment Cost is not limited: just for fun(?)

```python
# exclude outcomes which are from Policy (Irrelevant at this moment)
outcomes = []
direction = ScalarOutcome.MINIMIZE

for dike in dike_model.function.dikelist:
    for entry in ['Expected Annual Damage',
#                   'Dike Investment Costs',
                  'Expected Number of Deaths']:
        o = ScalarOutcome('{}_{}'.format(dike, entry),
                         kind = direction
                         )
        outcomes.append(o)
dike_model.outcomes.clear()
dike_model.outcomes = outcomes
```

```python
# very first
mul = 5

epsilon = [1e3, 0.0001] * mul
nfe = 100

convergence_metrics = [EpsilonProgress(),
                      ]
# Units of epsilon and hyper volume are determined by considering the Scale of Damage, Number of Death

start = time.time()
with MultiprocessingEvaluator(dike_model) as evaluator:
    results, convergence = evaluator.optimize(nfe = nfe, searchover = 'levers', 
                                              epsilons = epsilon,
                                              convergence = convergence_metrics, 
                                              reference = ref_scenario)
end = time.time()
print(end - start)
```

```python
# before speed up
mul = 5

epsilon = [1e4, 1e4] * mul
nfe = 100

convergence_metrics = [EpsilonProgress(),
                      ]
# Units of epsilon and hyper volume are determined by considering the Scale of Damage, Number of Death

start = time.time()
with MultiprocessingEvaluator(dike_model) as evaluator:
    results, convergence = evaluator.optimize(nfe = nfe, searchover = 'levers', 
                                              epsilons = epsilon,
                                              convergence = convergence_metrics, 
                                              reference = ref_scenario)
end = time.time()
print(end - start)
```

##### Dike Investment Cost is included

```python
# exclude outcomes which are from Policy (Irrelevant at this moment)
outcomes = []
direction = ScalarOutcome.MINIMIZE

for dike in dike_model.function.dikelist:
    for entry in ['Expected Annual Damage',
                  'Dike Investment Costs',
                  'Expected Number of Deaths']:
        o = ScalarOutcome('{}_{}'.format(dike, entry),
                         kind = direction
                         )
        outcomes.append(o)
outcomes.append(ScalarOutcome('RfR Total Costs', kind = direction))
outcomes.append(ScalarOutcome('Expected Evacuation Costs', kind = direction))

dike_model.outcomes.clear()
dike_model.outcomes = outcomes
```

```python
mul = 5

# Damage, Cost, Death
epsilon = [1e3, 1e3, 0.0001] * mul
nfe = 100

convergence_metrics = [EpsilonProgress(),
                      ]
# Units of epsilon and hyper volume are determined by considering the Scale of Damage, Number of Death

start = time.time()
with MultiprocessingEvaluator(dike_model) as evaluator:
    results, convergence = evaluator.optimize(nfe = nfe, searchover = 'levers', 
                                              epsilons = epsilon,
                                              convergence = convergence_metrics, 
                                              reference = ref_scenario)
end = time.time()
print(end - start)
```

### Reference Scenarios + 10,000 policies + 8 objectives*
#### we obtained 4 conclusions:

- ① **Dike Increase and EWS** does have a positive impact on reducing deaths.
- ② The effectiveness of **RfR** is not as significant as other levers.
- ③ **Trade-off between** objectives can be implicitly observed by **negative correlation** between **Death and Cost**
- ④ Strong **Correlation between Deaths and Damages** still holds, after introducing EWS lever.

```python
import time, itertools
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from custom_funcs import *

from ema_workbench import (Model, IntegerParameter, RealParameter, CategoricalParameter, Constant,
                           ScalarOutcome, MultiprocessingEvaluator, Policy, Scenario)

from ema_workbench.em_framework.evaluators import perform_experiments
from ema_workbench.em_framework.samplers import sample_uncertainties
from ema_workbench.em_framework.optimization import EpsilonProgress, HyperVolume
from ema_workbench.util import ema_logging, save_results, load_results
from ema_workbench.analysis import prim, feature_scoring

from problem_formulation import get_model_for_problem_formulation
ema_logging.log_to_stderr(ema_logging.INFO)
```

```python
# Problem Formulation for Everything (8 objectives)
dike_model = get_model_for_problem_formulation(5)
```

```python
dike_model.outcomes.keys()
```

```python
reference_values = {'Bmax': 175, 'Brate': 1.5, 'pfail': 0.5,
                    'discount rate': 3.5,
                    'ID flood wave shape': 4}
scen1 = {}
for key in dike_model.uncertainties:
        name_split = key.name.split('_')
        if len(name_split) == 1:
             scen1.update({key.name: reference_values[key.name]})
        else:
             scen1.update({key.name: reference_values[name_split[1]]})
ref_scenario = Scenario('reference', **scen1)
```

```python
n_policies = 10000

start = time.time()
with MultiprocessingEvaluator(dike_model) as evaluator:
    results = evaluator.perform_experiments(scenarios=ref_scenario, policies = n_policies)
    end = time.time()
print(end - start, "secs")
save_results(results, "1_4.tar.gz")
```

### 4) Impact of EWS
- on the Correlation between Deaths and Damages


*The coupled behavior of Damages and Deaths can be considerably decoupled, if EWS's effect is large.*

#### Fix EWS to 0

```python
dike_lev = {'DikeIncrease': [0, 10]}     # dm
rfr_lev = ['{}_RfR'.format(project_id) for project_id in range(0, 5)]
EWS_lev = {'EWS_DaysToThreat': [0, 0.01]}  # days

levers = []
for dike in ["A.1", "A.2", "A.3", "A.4", "A.5"]:
    for lev_name in dike_lev.keys():
        name = "{}_{}".format(dike, lev_name)
        levers.append(IntegerParameter(name, dike_lev[lev_name][0],
                                             dike_lev[lev_name][1]))
    for lev_name in rfr_lev:
        levers.append(IntegerParameter(lev_name, 0, 1))

for lev_name in EWS_lev.keys():
    levers.append(IntegerParameter(lev_name, EWS_lev[lev_name][0],
                                             EWS_lev[lev_name][1]))
```

```python
# Problem Formulation for Everything (17 objectives)
dike_model = get_model_for_problem_formulation(3)
```

```python
reference_values = {'Bmax': 175, 'Brate': 1.5, 'pfail': 0.5,
                    'discount rate': 3.5,
                    'ID flood wave shape': 4}
scen1 = {}
for key in dike_model.uncertainties:
        name_split = key.name.split('_')
        if len(name_split) == 1:
             scen1.update({key.name: reference_values[key.name]})
        else:
             scen1.update({key.name: reference_values[name_split[1]]})
ref_scenario = Scenario('reference', **scen1)
```

```python
n_policies = 5000

start = time.time()
with MultiprocessingEvaluator(dike_model) as evaluator:
    results = evaluator.perform_experiments(scenarios=ref_scenario, policies = n_policies)
    end = time.time()
print(end - start, "secs")
save_results(results, "1_6.tar.gz")
```

```python
results = load_results("1_6.tar.gz")
```

```python
import time
import itertools
import numpy as np
import networkx as nx
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from ema_workbench import (Model, ScalarOutcome, MultiprocessingEvaluator, Policy, Scenario)

from ema_workbench.em_framework.evaluators import perform_experiments, BaseEvaluator
from ema_workbench.em_framework.samplers import sample_uncertainties
from ema_workbench.em_framework.optimization import EpsilonProgress, HyperVolume, ArchiveLogger

from ema_workbench.util import ema_logging, save_results, load_results

from ema_workbench.analysis import prim, pairs_plotting, parcoords

from problem_formulation import get_model_for_problem_formulation
ema_logging.log_to_stderr(ema_logging.INFO)
```

```python
dike_model = get_model_for_problem_formulation(5)
```

```python
results = load_results("./results/8obj4pol_results.tar.gz")
experiments, outcomes = results
outcomes = pd.DataFrame(outcomes)
```

```python
frac_names = np.repeat("fraction ", outcomes.shape[1]).astype(object) + outcomes.columns
var_names = outcomes.columns
```

```python
q=0.9
thresholds = outcomes.quantile(q).to_dict()
thresholds
```

```python
import functools

def robustness(threshold, data):
    return np.sum(data<=threshold)/data.shape[0]

dict_funcs = {}
for key, thresh in thresholds.items():
    dict_funcs[key] = functools.partial(robustness, thresh)
dict_funcs
```

### Specify Rubustness Functions as Scalar Outcomes of EMA Workbench

```python
direction = ScalarOutcome.MAXIMIZE
robustness_functions = []
for i, var_name in enumerate(var_names):
    frac_name = "frac_{}".format(var_name)
    func = dict_funcs[var_name]
    
    robust_func = ScalarOutcome(
        name=frac_name,
        kind=direction,
        variable_name=var_name,
        function=func)
    
    robustness_functions.append(robust_func)
robustness_functions
```

### Run Optimization

```python
nfe = 1000
n_scenarios = 50
n_out = len(robustness_functions)
scenarios = sample_uncertainties(dike_model, n_scenarios)

decision_varnames = ["frac_{}".format(var_name) for var_name in var_names]
outcome_varnames = list(dike_model.levers.keys())

convergence = [
    EpsilonProgress(),
    HyperVolume(minimum=[0.0]*n_out, maximum=[1.01]*n_out),
    ArchiveLogger("./results_MORO",
                  decision_varnames=decision_varnames,
                  outcome_varnames=outcome_varnames)
               ]

epsilons = [1e-02]*n_out

start = time.time()
with MultiprocessingEvaluator(dike_model) as evaluator:
    results = evaluator.robust_optimize(robustness_functions, scenarios,
                                                     nfe=nfe,epsilons=epsilons,
                                                     convergence=convergence,
                                                     )
end = time.time()
print("Time:", round(end-start,2)/3600, "Hours")
```

```python
q = 0.9
fail = (outcomes > outcomes.quantile(q)).apply(lambda x: "fail" if any(x[:2]) else 0, axis = 1)
inputs["fail"] = fail.values
```

```python
sns.countplot("fail", data=inputs)
```

```python
params = list(dike_model.uncertainties.keys())
params.append("policy")
params.append("fail")
```

```python
import itertools

ncols=2
nrows=6
fig, axes = plt.subplots(ncols = ncols, nrows = nrows,
#                          sharex = True,
                         figsize = (5*ncols,5*nrows))
locs = list(itertools.product(range(nrows),range(ncols)))

params = list(dike_model.uncertainties.keys())
params = [param for param in params if ("Brate" not in param) & ("discount" not in param)]

for i,param in enumerate(params):
    loc=locs[i]
    ax=axes[loc]
    sns.violinplot("fail", param, data = inputs, ax=ax, legend=True,
                   hue="policy"
                  )
    ax.set_xlabel("")
    ax.set_xticklabels(inputs["fail"].unique(),fontsize=15)
    ax.set_ylabel("")
    ax.set_title(param)
fig.tight_layout()
fig.suptitle("How Policy can fail/success under varying Uncertainties", y=1.02, fontsize = 15)
```

#### Vulnerability for Categorical Uncertainties

```python
params = list(dike_model.uncertainties.keys())
params = [param for param in params if ("Brate" in param) | ("discount" in param)]
for param in params:
    g = sns.factorplot(x=param, hue="fail", col="policy", data=inputs, kind="count", palette="coolwarm")
    g.set_xlabels("")
    g.fig.suptitle(param, y=1.05, fontsize=15)
```

```python
import time
import numpy as np
import networkx as nx
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from ema_workbench import (Model, MultiprocessingEvaluator, ScalarOutcome, Policy, Scenario)

from ema_workbench.em_framework.evaluators import perform_experiments
from ema_workbench.em_framework.samplers import sample_uncertainties
from ema_workbench.em_framework.optimization import EpsilonProgress, HyperVolume, ArchiveLogger
from ema_workbench.analysis import parcoords

from ema_workbench.util import ema_logging

from problem_formulation import get_model_for_problem_formulation
ema_logging.log_to_stderr(ema_logging.INFO)
```

### 2. Search for a set of Optimal Solutions under Ref_Scenario

```python
dike_model = get_model_for_problem_formulation(5)

# 1. Determine a set of certain values
reference_values = {'Bmax': 175,
                    'Brate': 1.5,
                    'pfail': 0.5, 
                    'discount rate': 3.5,
                    'ID flood wave shape': 4}

# 2. Set the values as a Reference Scenario
scen1 = {}
for key in dike_model.uncertainties:
    name_split = key.name.split('_')
    if len(name_split) == 1:
        scen1.update({key.name: reference_values[key.name]})
    else:
        scen1.update({key.name: reference_values[name_split[1]]})

ref_scenario = Scenario('reference', **scen1)
```

```python
dike_model.outcomes.keys()
```

```python
rand_policy = {'DikeIncrease': 0, 'DaysToThreat': 0, 'RfR': 0}

pol0 = {}
for key in dike_model.levers:
    s1, s2 = key.name.split('_')
    pol0.update({key.name: rand_policy[s2]})                    
policy0 = Policy('Policy 0', **pol0)

start = time.time()
# single run with reference scenario and zero policy
dike_model.run_model(ref_scenario, policy0)
results = dike_model.outcomes_output
end = time.time()
print(end - start)
print(results)
```

```python
dike_model = get_model_for_problem_formulation(5)
rand_policy = {'DikeIncrease': 1, 'DaysToThreat': 2, 'RfR': 1}

pol0 = {}
for key in dike_model.levers:
    s1, s2 = key.name.split('_')
    pol0.update({key.name: rand_policy[s2]})                    
policy0 = Policy('Policy 0', **pol0)

start = time.time()
# single run with reference scenario and zero policy
dike_model.run_model(ref_scenario, policy0)
results = dike_model.outcomes_output
end = time.time()
print(end - start)
print(results)
```

```python
n_outcomes = len(results.values())
ref_value_outcomes = np.array(list(results.values()))
ref_value_outcomes
```

```python
epsilon = ref_value_outcomes / 10
epsilon
```

```python
decision_varnames=list(dike_model.levers.keys())
outcome_varnames=list(dike_model.outcomes.keys())

convergence_metrics = [
    EpsilonProgress(),
    HyperVolume(minimum=[0,]*n_outcomes, maximum=ref_value_outcomes*2),
    ArchiveLogger("./archive", decision_varnames=decision_varnames, outcome_varnames=outcome_varnames)
]

nfe = 200
start = time.time()
with MultiprocessingEvaluator(dike_model) as evaluator:
    results, convergence = evaluator.optimize(nfe = nfe, searchover = 'levers', 
                                              epsilons = epsilon, convergence = convergence_metrics, 
                                              reference = ref_scenario)
end = time.time()
print("Time:", round(end-start)/60, "Minutes")
```

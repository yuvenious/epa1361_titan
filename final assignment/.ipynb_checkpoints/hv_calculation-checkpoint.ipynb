```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from pygmo import hypervolume
```

### Convergence of Searching Levers under Worst Case

```python
hvs = []
for i in range(2,52):
    path = "./archive/worst/archive_"+str(i)+".csv"
    points = pd.read_csv(path, index_col=0).iloc[:, -8:].values
    ref_point = np.array([2,2,2,1e8,1e8,1e8,1e9,1e5])*1.2
    hv = hypervolume(points)
    hv = hv.compute(ref_point)
    hvs.append(hv)
plt.plot(hvs)
```

```python
hvs = []

for i in range(2,52):
    path = "./archive/MORO/6. final/1. 0.01death_CostMean/archive_"+str(i)+".csv"
    points = pd.read_csv(path, index_col=0).iloc[:, -8:].values
    points = points.tolist()
    ref_point = (np.ones(8) + 1)
    hv = hypervolume(points)
    hvc = hv.compute(ref_point)
    hvs.append(hvc)
plt.plot(hvs)
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from pygmo import hypervolume
```

### 1) Convergence of Searching Levers under Worst Case

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
pd.read_csv("2_1_con.csv")["hypervolume"].plot()
plt.title("Searching Worst Case under Zero-policy")
plt.xlabel("number of generation")
plt.ylabel("hyper-volume")
plt.grid()

plt.subplot(1,2,2)
hvs = [0]
ref_point = np.array([2,2,2,1e8,1e8,1e8,1e9,1e5])*1.14

for i in range(2,52):
    path = "./archive/worst/archive_"+str(i)+".csv"
    points = pd.read_csv(path, index_col=0).iloc[:, -8:].values
    hv = hypervolume(points)
    hv = hv.compute(ref_point)
    hvs.append(hv)
plt.plot(hvs)
plt.title("Searching Levers under the Worst Case")
plt.xlabel("number of generation")
plt.ylabel("hyper-volume")
plt.grid()
```

### 2) Convergence of MORO1 and MORO2

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

hvs = [0]
ref_point = (np.ones(8)*5)

for i in range(2,52):
    path = "./archive/MORO/6. final/1. 0.01death_CostMean/archive_"+str(i)+".csv"
    points = pd.read_csv(path, index_col=0).iloc[:, -8:].values
    points = points.tolist()
    hv = hypervolume(points)
    hvc = hv.compute(ref_point)
    hvs.append(hvc)
plt.plot(hvs)
plt.xlabel("number of generation")
plt.ylabel("hyper-volume")
plt.grid()
plt.title("MORO1")

plt.subplot(1,2,2)
hvs = [0]
ref_point = (np.ones(8)*5)

for i in range(2,52):
    path = "./archive/MORO/6. final/2. 0death_Cost90%/archive_"+str(i)+".csv"
    points = pd.read_csv(path, index_col=0).iloc[:, -8:].values
    points = points.tolist()
    hv = hypervolume(points)
    hvc = hv.compute(ref_point)
    hvs.append(hvc)
plt.plot(hvs)
plt.xlabel("number of generation")
plt.ylabel("hyper-volume")
plt.grid()
plt.title("MORO2")
```

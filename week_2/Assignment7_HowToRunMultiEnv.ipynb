#### Before Policy

```python
n_iter = 100

main = ['infected population region 1', 'infected population region 2',
        'recovered population region 1', 'recovered population region 2']

models = pd.DataFrame(data=[[itr, pysd.read_vensim('flumdl.mdl')] for itr in range(n_iter)],
                      columns=['iter', 'model'])
models["contact rate region 1"] = np.random.randint(25, 75, n_iter)
models["contact rate region 2"] = np.random.randint(75, 125, n_iter)
print("models created")

for index, row in models.iterrows():
    row['model'].set_components({'contact rate region 1':row['contact rate region 1'],
                                 'contact rate region 2':row['contact rate region 2']})

print("params are set and start plotting")

# plot
fig, axes = plt.subplots(ncols = 2, nrows = 2, figsize = (15, 7))
location = (0,0), (0,1), (1,0), (1,1)

for index, row in models.iterrows():
    if (index !=0) & (index%10 == 0):
        print(index)
    df = row['model'].run()
    for i in range(len(main)):
        axes[location[i]].plot(df[main[i]])

# setting details
for i in range(len(main)):
    axes[location[i]].set_title(main[i])
#     axes[location[i]].set_xlabel("months")
```

#### After Policy (Disconnect to regions)

```python
n_iter = 100

main = ['infected population region 1', 'infected population region 2',
        'recovered population region 1', 'recovered population region 2']

models = pd.DataFrame(data=[[itr, pysd.read_vensim('flumdl.mdl')] for itr in range(n_iter)],
                      columns=['iter', 'model'])
models["contact rate region 1"] = np.random.randint(25, 75, n_iter)
models["contact rate region 2"] = np.random.randint(75, 125, n_iter)
print("models created")

for index, row in models.iterrows():
    row['model'].set_components({'contact rate region 1':row['contact rate region 1'],
                                 'contact rate region 2':row['contact rate region 2'],
# POLICY IMPLEMENTATION!
                                 'switch': 0})

print("params are set and start plotting")

# plot
fig, axes = plt.subplots(ncols = 2, nrows = 2, figsize = (15, 7))
location = (0,0), (0,1), (1,0), (1,1)

for index, row in models.iterrows():
    if (index !=0) & (index%10 == 0):
        print(index)
    df = row['model'].run()
    for i in range(len(main)):
        axes[location[i]].plot(df[main[i]])

# setting details
for i in range(len(main)):
    axes[location[i]].set_title(main[i])
#     axes[location[i]].set_xlabel("months")
```

## EMA workbench

```python
from ema_workbench import (TimeSeriesOutcome, perform_experiments, Policy,
                           RealParameter, ema_logging)
```

```python
# from ema_workbench import 
from ema_workbench.connectors.vensim import VensimModel
from ema_workbench.connectors import pysd_connector
ema_logging.log_to_stderr(ema_logging.INFO)
```

```python
# model = VensimModel(name = "model", wd = os.getcwd(), model_file="flumdl.vpm")
# EMA workbench dosen't work for now.

model = pysd_connector.PysdModel(name = "mdl", mdl_file = "flumdl.mdl")
```

### Befor Policy

```python
model.uncertainties = [
    RealParameter('contact rate region 1', 25 , 75),
    RealParameter('contact rate region 2', 75 , 125),
    RealParameter('susceptible to immune population delay time region 1', 0, 2),
    RealParameter('susceptible to immune population delay time region 2', 0, 2),
    RealParameter('recovering time', 0.1, 2),
    RealParameter('fatality ratio', 0.001, 0.2),
    RealParameter('infection rate region 1', 0.001, 0.2),
    RealParameter('infection rate region 2', 0.001, 0.2),    
]
    
model.outcomes = [
    TimeSeriesOutcome('infected population region 1'),
    TimeSeriesOutcome('infected population region 2'),
    TimeSeriesOutcome('recovered population region 1'),
    TimeSeriesOutcome('recovered population region 2'),
]
```

```python
n_scenarios = 100
result = perform_experiments(models = model, scenarios = n_scenarios)
```
